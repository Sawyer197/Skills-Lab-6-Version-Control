```python
import random
print("Welcome to Math Quiz!\nYou will be tested on your mathematical prowess for 5 questions")
numAnswers=0
numCorrect=0
while(numAnswers<5):
    correctAnswer=0.0
    operationString=""
    operation=random.randint(0,3)
    num1=float(random.randint(-10.0,10.0))
    num2=float(random.randint(-10.0,10.0))
    while(num2==0 and operation==3):
        num2=float(random.randint(-10.0,10.0))
    validInput=False
    while(validInput!=True):
        try:
            operationString
            if(operation==0):
                operationString="+"
            if(operation==1):
                operationString="-"
            if(operation==2):
                operationString="*"
            if(operation==3):
                operationString="/"
            question=str(num1)+operationString+str(num2)
            answer=float(input("Try:\n"+question+"="))
            validInput=True
        except:
            print("Invalid input!")
    correctAnswer=0
    exec("correctAnswer="+question)
    if(answer==correctAnswer):
        numCorrect+=1
        print("Correct!")
    else:
        print("wrong.")
    numAnswers+=1
print("You got",numCorrect,"questions right!
```
